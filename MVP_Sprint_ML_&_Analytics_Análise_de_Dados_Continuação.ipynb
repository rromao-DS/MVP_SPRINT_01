# Funções Auxiliares


Crei essa categoria para ajudar a modularizar os códigos e torná-los mais fáceis de entender, manter e reutilizar. Em vez de escrever todo o código dentro de um único bloco, eu dividi o código em algumas funções menores para realizar tarefas específicas.

**cross_validation:** Esta função é essencial para avaliar o desempenho de modelos de machine learning. A validação cruzada k-fold é uma técnica amplamente aceita que permite estimar a performance de um modelo em dados não vistos. Isso é importante para evitar overfitting e para entender como o modelo se comporta em diferentes subconjuntos de dados. A função cross_validation executa a validação cruzada k-fold, treinando o modelo em diferentes subconjuntos de dados e calculando métricas de desempenho (MAE, MAPE e RMSE) para cada fold. Isso fornece uma estimativa mais confiável do desempenho do modelo.

**mean_percentage_error e mean_absolute_percentage_error:** Essas funções são usadas para calcular duas métricas importantes de erro de regressão: MAE (Mean Absolute Error) e MAPE (Mean Absolute Percentage Error). O MAE mede o erro médio absoluto entre as previsões do modelo e os valores reais. O MAPE é uma medida adicional que expressa o erro médio em termos percentuais, facilitando a interpretação do desempenho do modelo. Ambas as métricas oferecem uma avaliação robusta do desempenho do modelo.

**ml_error:** Esta função calcula várias métricas de erro de regressão (MAE, MAPE e RMSE) entre as previsões do modelo e os valores reais. O MAE e o RMSE fornecem medidas de erro absolutas e relativas, enquanto o MAPE fornece uma medida adicional em termos percentuais. Ao apresentar todas essas métricas em um DataFrame pandas, é possível realizar uma comparação abrangente do desempenho de diferentes modelos.

**cramer_v:** Esta função calcula o coeficiente de associação de Cramer's V entre duas variáveis categóricas. É muito importante explorar a associação entre variáveis do conjunto de dados que vou apresentar. O coeficiente de Cramer's V é uma medida comum de associação para variáveis categóricas e fornece informações sobre a força e direção da associação. Isso é útil para análises exploratórias e para entender a relação entre as variáveis do estudo.

```python
def cross_validation(x_training, kfold, model_name, model, verbose=False):
    mae_list = []
    mape_list = []
    rmse_list = []

    for k in reversed(range(1, kfold + 1)):
        if verbose:
            print('\nKFold Number: {}'.format(k))

        # Definir as datas de início e fim para a validação
        validation_start_date = x_training['date'].max() - datetime.timedelta(days=k * 6 * 7)
        validation_end_date = x_training['date'].max() - datetime.timedelta(days=(k - 1) * 6 * 7)

        # Filtrar o conjunto de dados
        training = x_training[x_training['date'] < validation_start_date]
        validation = x_training[(x_training['date'] >= validation_start_date) & (x_training['date'] <= validation_end_date)]

        # Remover linhas com valores NaN dos dados de treinamento e validação
        training_clean = training.dropna()
        validation_clean = validation.dropna()

        # Verificar se ainda há dados após a remoção dos NaNs
        if training_clean.shape[0] == 0 or validation_clean.shape[0] == 0:
            continue

        # Dados de treinamento e validação
        x_training_clean = training_clean.drop(['date', 'sales'], axis=1)
        y_training_clean = training_clean['sales']
        x_validation_clean = validation_clean.drop(['date', 'sales'], axis=1)
        y_validation_clean = validation_clean['sales']

        # Treinar o modelo
        model.fit(x_training_clean, y_training_clean)

        # Fazer previsões
        yhat = model.predict(x_validation_clean)

        # Calcular as métricas de erro
        m_result = ml_error(model_name, np.expm1(y_validation_clean), np.expm1(yhat))

        # Armazenar as métricas de erro de cada iteração do kfold
        mae_list.append(m_result['MAE'])
        mape_list.append(m_result['MAPE'])
        rmse_list.append(m_result['RMSE'])

    # Calcular a média e desvio padrão das métricas de erro
    mae_mean = np.mean(mae_list)
    mae_std = np.std(mae_list)
    mape_mean = np.mean(mape_list)
    mape_std = np.std(mape_list)
    rmse_mean = np.mean(rmse_list)
    rmse_std = np.std(rmse_list)

    # Criar DataFrame com as métricas de erro médias e desvios padrão
    result_df = pd.DataFrame({
        'Model Name': model_name,
        'MAE CV': f'{mae_mean:.2f} +/- {mae_std:.2f}',
        'MAPE CV': f'{mape_mean:.2f} +/- {mape_std:.2f}',
        'RMSE CV': f'{rmse_mean:.2f} +/- {rmse_std:.2f}'
    }, index=[0])

    return result_df


def mean_percentage_error( y, yhat ):
    return np.mean( ( y - yhat ) / y )


def mean_absolute_percentage_error( y, yhat ):
    return np.mean( np.abs( ( y - yhat ) / y ) )


def ml_error( model_name, y, yhat ):
    mae = mean_absolute_error( y, yhat )
    mape = mean_absolute_percentage_error( y, yhat )
    rmse = np.sqrt( mean_squared_error( y, yhat ) )

    return pd.DataFrame( { 'Model Name': model_name,
                           'MAE': mae,
                           'MAPE': mape,
                           'RMSE': rmse }, index=[0] )

def cramer_v( x, y ):
    cm = pd.crosstab( x, y ).to_numpy()
    n = cm.sum()
    r, k = cm.shape

    chi2 = ss.chi2_contingency( cm )[0]
    chi2corr = max( 0, chi2 - (k-1)*(r-1)/(n-1) )

    kcorr = k - (k-1)**2/(n-1)
    rcorr = r - (r-1)**2/(n-1)

    return np.sqrt( (chi2corr/n) / ( min( kcorr-1, rcorr-1 ) ) )

```

## 2.4. Feature Engineering

Nesse momento é o processo de criação de novas variáveis com base nas variáveis existentes em no dataset que carreguei acima. Entendi que a partir da variável date poderia trazer novas features que seriam válidas para análise e maior conhecimento do negócio.

Então, criei aqui, a partir da feature ['date'] as seguintes colunas:

- Year;
- Month;
- day;
- Week_of_year (contendo o número da semana do ano) que retornará um valor numérico que poderei realizar operações matemáticas e cálculos;
- Year_week (contendo o ano e o número da semana concatenados) retornará uma string que poderei utiliar para fazer visualização e para agrupar dados por período;
- Competition_Since indicará a data em que cada loja começou a competir com a loja concorrente mais próxima;
- Competition_Time_Month indicará o número de meses desde o início da competição com a loja concorrente mais próxima.
- Promo_Since indicará a semana em que cada loja começou a participar da promoção Promo2.
- Promo_Time_Week indicará o número de semanas desde o início da promoção Promo2.

Além disso, nesse snippet temos algumas coversões ocorrendo no código como no assortment, promo_since e etc.

```python
# year
df2['year'] = df2['date'].dt.year

# month
df2['month'] = df2['date'].dt.month

# day
df2['day'] = df2['date'].dt.day

# week of year
df2['week_of_year'] = df2['date'].dt.isocalendar().week


# year week
df2['year_week'] = df2['date'].dt.strftime( '%Y-%W' )

# competition since
df2['competition_since'] = df2.apply( lambda x: datetime.datetime( year=x['competition_open_since_year'], month=x['competition_open_since_month'],day=1 ), axis=1 )

df2['competition_time_month'] = ( ( df2['date'] - df2['competition_since'] )/30 ).apply( lambda x: x.days )
df2['competition_time_month'] = df2['competition_time_month'].fillna(df2['competition_time_month'].median().astype(int))

# promo since
df2['promo_since'] = df2['promo2_since_year'].astype( str ) + '-' + df2['promo2_since_week'].astype( str )
df2['promo_since'] = df2['promo_since'].apply( lambda x: datetime.datetime.strptime( x + '-1', '%Y-%W-%w' ) - datetime.timedelta( days=7 ) )
df2['promo_time_week'] = ( ( df2['date'] - df2['promo_since'] )/7 ).apply( lambda x: x.days )
df2['promo_time_week'] = df2['promo_time_week'].fillna(df2['promo_time_week'].median().astype( int ))

# assortment
df2['assortment'] = df2['assortment'].apply( lambda x: 'basic' if x == 'a' else 'extra' if x == 'b' else 'extended' )

# state holiday
df2['state_holiday'] = df2['state_holiday'].apply( lambda x: 'public_holiday' if x == 'a' else 'easter_holiday' if x == 'b' else 'christmas' if x == 'c' else 'regular_day' )
```

### 5.4.1 Encoding

Essa etapa de pré-processamento de dado é essencial para garantir que os dados estejam em um formato adequado para a modelagem e que o modelo final seja capaz de aprender e generalizar a partir dos dados de forma eficaz.

- One Hot Encoding: é usada para converter variáveis categóricas em uma representação numérica, criando colunas binárias para cada categoria única presente na variável. Isso permite que o modelo aprenda a importância de cada categoria sem assumir uma relação ordinal entre elas.
- Label Encoding: Esta técnica é usada quando há uma ordem natural nas categorias da variável categórica. Cada categoria única é mapeada para um valor numérico inteiro. Isso simplifica a representação dos dados, mas pode introduzir uma ordem artificial entre as categorias.
- Ordinal Encoding: Similar ao Label Encoding, esta técnica é usada quando há uma ordem específica nas categorias da variável categórica. Cada categoria é mapeada para um valor numérico de acordo com a ordem especificada, preservando a relação ordinal entre as categorias.


```python
# state_holiday - One Hot Encoding
df5 = pd.get_dummies( df5, prefix=['state_holiday'], columns=['state_holiday'] )

# store_type - Label Encoding
le = LabelEncoder()
df5['store_type'] = le.fit_transform( df5['store_type'] )

# assortment - Ordinal Encoding
assortment_dict = {'basic': 1,  'extra': 2, 'extended': 3}
df5['assortment'] = df5['assortment'].map( assortment_dict )
```

### 5.4.2. Transformação da Variável Resposta

```python
df5['sales'] = np.log1p( df5['sales'] )
```

Ao aplicar a transformação logarítmica aos dados de vendas que é a nossa variavel resposta, estou ajustando a distribuição dos dados para melhorar a precisão e interpretabilidade da previsão de vendas futuras.

### 5.4.3 Transformação da Natureza

- Nesta etapa, estou transformando a natureza dos dados temporais para capturar padrões sazonais e periódicos de uma maneira mais eficaz. Para isso, estou criando variáveis cíclicas que representam o dia da semana, mês, dia do mês e semana do ano. Essas variáveis são fundamentais para permitir que o modelo de machine learning compreenda e capture os padrões temporais presentes nos dados de vendas.

**Variáveis para o Dia da Semana (day_of_week_sin e day_of_week_cos):**
Essas variáveis capturam a periodicidade semanal dos dados, permitindo que o modelo identifique padrões que se repetem a cada semana.

**Variáveis para o Mês (month_sin e month_cos):**
Representam a sazonalidade mensal dos dados, possibilitando a detecção de padrões que variam ao longo do ano.

**Variáveis para o Dia do Mês (day_sin e day_cos):**
Permitem capturar padrões que ocorrem em intervalos regulares ao longo do mês, como variações no início e no final do mês.

**Variáveis para a Semana do Ano (week_of_year_sin e week_of_year_cos):**
Essas variáveis ajudam o modelo a identificar padrões sazonais de longo prazo que ocorrem em diferentes semanas do ano.

Essa transformação é crucial para garantir que o modelo seja capaz de capturar adequadamente os padrões temporais nos dados de vendas, melhorando assim a qualidade das previsões futuras.

```python
# day of week
df5['day_of_week_sin'] = df5['day_of_week'].apply( lambda x: np.sin( x * ( 2. * np.pi/7 ) ) )
df5['day_of_week_cos'] = df5['day_of_week'].apply( lambda x: np.cos( x * ( 2. * np.pi/7 ) ) )

# month
df5['month_sin'] = df5['month'].apply( lambda x: np.sin( x * ( 2. * np.pi/12 ) ) )
df5['month_cos'] = df5['month'].apply( lambda x: np.cos( x * ( 2. * np.pi/12 ) ) )

# day
df5['day_sin'] = df5['day'].apply( lambda x: np.sin( x * ( 2. * np.pi/30 ) ) )
df5['day_cos'] = df5['day'].apply( lambda x: np.cos( x * ( 2. * np.pi/30 ) ) )

# week of year
df5['week_of_year_sin'] = df5['week_of_year'].apply( lambda x: np.sin( x * ( 2. * np.pi/52 ) ) )
df5['week_of_year_cos'] = df5['week_of_year'].apply( lambda x: np.cos( x * ( 2. * np.pi/52 ) ) )
```

## 7.1. Modelo de Média

Estou considerando a implementação de um modelo de média. Esta abordagem é bastante simples, mas muitas vezes eficaz, especialmente como uma linha de base para comparação com modelos mais complexos. Ele calcula a média dos valores históricos e utiliza esse valor como a previsão para o próximo período. Ele não leva em consideração padrões sazonais, tendências ou outras estruturas nos dados. No entanto, pode ser útil para fornecer uma estimativa inicial ou como um ponto de referência para avaliar a performance de modelos mais sofisticados.

Na implementação prática, planejo usar bibliotecas como pandas e numpy em Python para calcular a média dos valores históricos. Em seguida, vou avaliar a eficácia desse modelo em comparação com outras abordagens mais avançadas que estou explorando. Isso me ajudará a entender melhor o desempenho relativo dos diferentes modelos e tomar decisões informadas sobre qual abordagem adotar para previsão de séries temporais.

```python
# Calcular as previsões do modelo de média
aux1 = x_test.copy()
aux1['sales'] = y_test.copy()

aux2 = aux1[['store', 'sales']].groupby('store').mean().reset_index().rename(columns={'sales': 'predictions'})
aux1 = pd.merge(aux1, aux2, how='left', on='store')
yhat_baseline = aux1['predictions']

# Preencher valores NaN nas previsões com a média global das vendas
mean_sales = yhat_baseline.mean()
yhat_baseline.fillna(mean_sales, inplace=True)

# Remover linhas com valores NaN dos dados de teste e das previsões
aux1.dropna(subset=['sales', 'predictions'], inplace=True)

# Calcular as métricas de erro apenas com os dados limpos
baseline_result = ml_error('Average Model', np.expm1(aux1['sales']), np.expm1(aux1['predictions']))
baseline_result
```

O MAPE de 3.335177% indica que o erro médio em relação às vendas reais é de cerca de 3.34%. Por fim, o RMSE de 0.46782 mostra que a dispersão dos erros é relativamente baixa.
