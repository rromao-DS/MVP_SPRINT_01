# Fun√ß√µes Auxiliares


Crei essa categoria para ajudar a modularizar os c√≥digos e torn√°-los mais f√°ceis de entender, manter e reutilizar. Em vez de escrever todo o c√≥digo dentro de um √∫nico bloco, eu dividi o c√≥digo em algumas fun√ß√µes menores para realizar tarefas espec√≠ficas.

**cross_validation:** Esta fun√ß√£o √© essencial para avaliar o desempenho de modelos de machine learning. A valida√ß√£o cruzada k-fold √© uma t√©cnica amplamente aceita que permite estimar a performance de um modelo em dados n√£o vistos. Isso √© importante para evitar overfitting e para entender como o modelo se comporta em diferentes subconjuntos de dados. A fun√ß√£o cross_validation executa a valida√ß√£o cruzada k-fold, treinando o modelo em diferentes subconjuntos de dados e calculando m√©tricas de desempenho (MAE, MAPE e RMSE) para cada fold. Isso fornece uma estimativa mais confi√°vel do desempenho do modelo.

**mean_percentage_error e mean_absolute_percentage_error:** Essas fun√ß√µes s√£o usadas para calcular duas m√©tricas importantes de erro de regress√£o: MAE (Mean Absolute Error) e MAPE (Mean Absolute Percentage Error). O MAE mede o erro m√©dio absoluto entre as previs√µes do modelo e os valores reais. O MAPE √© uma medida adicional que expressa o erro m√©dio em termos percentuais, facilitando a interpreta√ß√£o do desempenho do modelo. Ambas as m√©tricas oferecem uma avalia√ß√£o robusta do desempenho do modelo.

**ml_error:** Esta fun√ß√£o calcula v√°rias m√©tricas de erro de regress√£o (MAE, MAPE e RMSE) entre as previs√µes do modelo e os valores reais. O MAE e o RMSE fornecem medidas de erro absolutas e relativas, enquanto o MAPE fornece uma medida adicional em termos percentuais. Ao apresentar todas essas m√©tricas em um DataFrame pandas, √© poss√≠vel realizar uma compara√ß√£o abrangente do desempenho de diferentes modelos.

**cramer_v:** Esta fun√ß√£o calcula o coeficiente de associa√ß√£o de Cramer's V entre duas vari√°veis categ√≥ricas. √â muito importante explorar a associa√ß√£o entre vari√°veis do conjunto de dados que vou apresentar. O coeficiente de Cramer's V √© uma medida comum de associa√ß√£o para vari√°veis categ√≥ricas e fornece informa√ß√µes sobre a for√ßa e dire√ß√£o da associa√ß√£o. Isso √© √∫til para an√°lises explorat√≥rias e para entender a rela√ß√£o entre as vari√°veis do estudo.

```python
def cross_validation(x_training, kfold, model_name, model, verbose=False):
    mae_list = []
    mape_list = []
    rmse_list = []

    for k in reversed(range(1, kfold + 1)):
        if verbose:
            print('\nKFold Number: {}'.format(k))

        # Definir as datas de in√≠cio e fim para a valida√ß√£o
        validation_start_date = x_training['date'].max() - datetime.timedelta(days=k * 6 * 7)
        validation_end_date = x_training['date'].max() - datetime.timedelta(days=(k - 1) * 6 * 7)

        # Filtrar o conjunto de dados
        training = x_training[x_training['date'] < validation_start_date]
        validation = x_training[(x_training['date'] >= validation_start_date) & (x_training['date'] <= validation_end_date)]

        # Remover linhas com valores NaN dos dados de treinamento e valida√ß√£o
        training_clean = training.dropna()
        validation_clean = validation.dropna()

        # Verificar se ainda h√° dados ap√≥s a remo√ß√£o dos NaNs
        if training_clean.shape[0] == 0 or validation_clean.shape[0] == 0:
            continue

        # Dados de treinamento e valida√ß√£o
        x_training_clean = training_clean.drop(['date', 'sales'], axis=1)
        y_training_clean = training_clean['sales']
        x_validation_clean = validation_clean.drop(['date', 'sales'], axis=1)
        y_validation_clean = validation_clean['sales']

        # Treinar o modelo
        model.fit(x_training_clean, y_training_clean)

        # Fazer previs√µes
        yhat = model.predict(x_validation_clean)

        # Calcular as m√©tricas de erro
        m_result = ml_error(model_name, np.expm1(y_validation_clean), np.expm1(yhat))

        # Armazenar as m√©tricas de erro de cada itera√ß√£o do kfold
        mae_list.append(m_result['MAE'])
        mape_list.append(m_result['MAPE'])
        rmse_list.append(m_result['RMSE'])

    # Calcular a m√©dia e desvio padr√£o das m√©tricas de erro
    mae_mean = np.mean(mae_list)
    mae_std = np.std(mae_list)
    mape_mean = np.mean(mape_list)
    mape_std = np.std(mape_list)
    rmse_mean = np.mean(rmse_list)
    rmse_std = np.std(rmse_list)

    # Criar DataFrame com as m√©tricas de erro m√©dias e desvios padr√£o
    result_df = pd.DataFrame({
        'Model Name': model_name,
        'MAE CV': f'{mae_mean:.2f} +/- {mae_std:.2f}',
        'MAPE CV': f'{mape_mean:.2f} +/- {mape_std:.2f}',
        'RMSE CV': f'{rmse_mean:.2f} +/- {rmse_std:.2f}'
    }, index=[0])

    return result_df


def mean_percentage_error( y, yhat ):
    return np.mean( ( y - yhat ) / y )


def mean_absolute_percentage_error( y, yhat ):
    return np.mean( np.abs( ( y - yhat ) / y ) )


def ml_error( model_name, y, yhat ):
    mae = mean_absolute_error( y, yhat )
    mape = mean_absolute_percentage_error( y, yhat )
    rmse = np.sqrt( mean_squared_error( y, yhat ) )

    return pd.DataFrame( { 'Model Name': model_name,
                           'MAE': mae,
                           'MAPE': mape,
                           'RMSE': rmse }, index=[0] )

def cramer_v( x, y ):
    cm = pd.crosstab( x, y ).to_numpy()
    n = cm.sum()
    r, k = cm.shape

    chi2 = ss.chi2_contingency( cm )[0]
    chi2corr = max( 0, chi2 - (k-1)*(r-1)/(n-1) )

    kcorr = k - (k-1)**2/(n-1)
    rcorr = r - (r-1)**2/(n-1)

    return np.sqrt( (chi2corr/n) / ( min( kcorr-1, rcorr-1 ) ) )

```

## 2.4. Feature Engineering

Nesse momento √© o processo de cria√ß√£o de novas vari√°veis com base nas vari√°veis existentes em no dataset que carreguei acima. Entendi que a partir da vari√°vel date poderia trazer novas features que seriam v√°lidas para an√°lise e maior conhecimento do neg√≥cio.

Ent√£o, criei aqui, a partir da feature ['date'] as seguintes colunas:

- Year;
- Month;
- day;
- Week_of_year (contendo o n√∫mero da semana do ano) que retornar√° um valor num√©rico que poderei realizar opera√ß√µes matem√°ticas e c√°lculos;
- Year_week (contendo o ano e o n√∫mero da semana concatenados) retornar√° uma string que poderei utiliar para fazer visualiza√ß√£o e para agrupar dados por per√≠odo;
- Competition_Since indicar√° a data em que cada loja come√ßou a competir com a loja concorrente mais pr√≥xima;
- Competition_Time_Month indicar√° o n√∫mero de meses desde o in√≠cio da competi√ß√£o com a loja concorrente mais pr√≥xima.
- Promo_Since indicar√° a semana em que cada loja come√ßou a participar da promo√ß√£o Promo2.
- Promo_Time_Week indicar√° o n√∫mero de semanas desde o in√≠cio da promo√ß√£o Promo2.

Al√©m disso, nesse snippet temos algumas covers√µes ocorrendo no c√≥digo como no assortment, promo_since e etc.

```python
# year
df2['year'] = df2['date'].dt.year

# month
df2['month'] = df2['date'].dt.month

# day
df2['day'] = df2['date'].dt.day

# week of year
df2['week_of_year'] = df2['date'].dt.isocalendar().week


# year week
df2['year_week'] = df2['date'].dt.strftime( '%Y-%W' )

# competition since
df2['competition_since'] = df2.apply( lambda x: datetime.datetime( year=x['competition_open_since_year'], month=x['competition_open_since_month'],day=1 ), axis=1 )

df2['competition_time_month'] = ( ( df2['date'] - df2['competition_since'] )/30 ).apply( lambda x: x.days )
df2['competition_time_month'] = df2['competition_time_month'].fillna(df2['competition_time_month'].median().astype(int))

# promo since
df2['promo_since'] = df2['promo2_since_year'].astype( str ) + '-' + df2['promo2_since_week'].astype( str )
df2['promo_since'] = df2['promo_since'].apply( lambda x: datetime.datetime.strptime( x + '-1', '%Y-%W-%w' ) - datetime.timedelta( days=7 ) )
df2['promo_time_week'] = ( ( df2['date'] - df2['promo_since'] )/7 ).apply( lambda x: x.days )
df2['promo_time_week'] = df2['promo_time_week'].fillna(df2['promo_time_week'].median().astype( int ))

# assortment
df2['assortment'] = df2['assortment'].apply( lambda x: 'basic' if x == 'a' else 'extra' if x == 'b' else 'extended' )

# state holiday
df2['state_holiday'] = df2['state_holiday'].apply( lambda x: 'public_holiday' if x == 'a' else 'easter_holiday' if x == 'b' else 'christmas' if x == 'c' else 'regular_day' )
```

### 5.4.1 Encoding

Essa etapa de pr√©-processamento de dado √© essencial para garantir que os dados estejam em um formato adequado para a modelagem e que o modelo final seja capaz de aprender e generalizar a partir dos dados de forma eficaz.

- One Hot Encoding: √© usada para converter vari√°veis categ√≥ricas em uma representa√ß√£o num√©rica, criando colunas bin√°rias para cada categoria √∫nica presente na vari√°vel. Isso permite que o modelo aprenda a import√¢ncia de cada categoria sem assumir uma rela√ß√£o ordinal entre elas.
- Label Encoding: Esta t√©cnica √© usada quando h√° uma ordem natural nas categorias da vari√°vel categ√≥rica. Cada categoria √∫nica √© mapeada para um valor num√©rico inteiro. Isso simplifica a representa√ß√£o dos dados, mas pode introduzir uma ordem artificial entre as categorias.
- Ordinal Encoding: Similar ao Label Encoding, esta t√©cnica √© usada quando h√° uma ordem espec√≠fica nas categorias da vari√°vel categ√≥rica. Cada categoria √© mapeada para um valor num√©rico de acordo com a ordem especificada, preservando a rela√ß√£o ordinal entre as categorias.


```python
# state_holiday - One Hot Encoding
df5 = pd.get_dummies( df5, prefix=['state_holiday'], columns=['state_holiday'] )

# store_type - Label Encoding
le = LabelEncoder()
df5['store_type'] = le.fit_transform( df5['store_type'] )

# assortment - Ordinal Encoding
assortment_dict = {'basic': 1,  'extra': 2, 'extended': 3}
df5['assortment'] = df5['assortment'].map( assortment_dict )
```

### 5.4.3 Transforma√ß√£o da Natureza

- Nesta etapa, estou transformando a natureza dos dados temporais para capturar padr√µes sazonais e peri√≥dicos de uma maneira mais eficaz. Para isso, estou criando vari√°veis c√≠clicas que representam o dia da semana, m√™s, dia do m√™s e semana do ano. Essas vari√°veis s√£o fundamentais para permitir que o modelo de machine learning compreenda e capture os padr√µes temporais presentes nos dados de vendas.

**Vari√°veis para o Dia da Semana (day_of_week_sin e day_of_week_cos):**
Essas vari√°veis capturam a periodicidade semanal dos dados, permitindo que o modelo identifique padr√µes que se repetem a cada semana.

**Vari√°veis para o M√™s (month_sin e month_cos):**
Representam a sazonalidade mensal dos dados, possibilitando a detec√ß√£o de padr√µes que variam ao longo do ano.

**Vari√°veis para o Dia do M√™s (day_sin e day_cos):**
Permitem capturar padr√µes que ocorrem em intervalos regulares ao longo do m√™s, como varia√ß√µes no in√≠cio e no final do m√™s.

**Vari√°veis para a Semana do Ano (week_of_year_sin e week_of_year_cos):**
Essas vari√°veis ajudam o modelo a identificar padr√µes sazonais de longo prazo que ocorrem em diferentes semanas do ano.

Essa transforma√ß√£o √© crucial para garantir que o modelo seja capaz de capturar adequadamente os padr√µes temporais nos dados de vendas, melhorando assim a qualidade das previs√µes futuras.

```python
# day of week
df5['day_of_week_sin'] = df5['day_of_week'].apply( lambda x: np.sin( x * ( 2. * np.pi/7 ) ) )
df5['day_of_week_cos'] = df5['day_of_week'].apply( lambda x: np.cos( x * ( 2. * np.pi/7 ) ) )

# month
df5['month_sin'] = df5['month'].apply( lambda x: np.sin( x * ( 2. * np.pi/12 ) ) )
df5['month_cos'] = df5['month'].apply( lambda x: np.cos( x * ( 2. * np.pi/12 ) ) )

# day
df5['day_sin'] = df5['day'].apply( lambda x: np.sin( x * ( 2. * np.pi/30 ) ) )
df5['day_cos'] = df5['day'].apply( lambda x: np.cos( x * ( 2. * np.pi/30 ) ) )

# week of year
df5['week_of_year_sin'] = df5['week_of_year'].apply( lambda x: np.sin( x * ( 2. * np.pi/52 ) ) )
df5['week_of_year_cos'] = df5['week_of_year'].apply( lambda x: np.cos( x * ( 2. * np.pi/52 ) ) )
```

Na implementa√ß√£o pr√°tica, planejo usar bibliotecas como pandas e numpy em Python para calcular a m√©dia dos valores hist√≥ricos. Em seguida, vou avaliar a efic√°cia desse modelo em compara√ß√£o com outras abordagens mais avan√ßadas que estou explorando.
